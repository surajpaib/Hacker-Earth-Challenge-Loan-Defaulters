```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Functions List

```python
def read_file(train_file):
    """
    Read csv file
    :param train_file:
    :return:
    """
    dataframe = pd.read_csv(train_file, delimiter=",")
    return dataframe


def remove_redundant_fields(dataframe, *args):
    """
    Remove Fields from Dataframe
    :param dataframe:
    :param args:
    :return:
    """
    for arg in args:
        del dataframe[arg]
        
    return dataframe
```

## Read the file with the training data set

```python
dataframe = read_file('Hacker Earth Challenge/train_indessa.csv')
```
